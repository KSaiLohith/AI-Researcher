```python

import os
from crewai import Agent, Task, Crew, Process
from langchain_groq import ChatGroq
from google.colab import userdata
```

```python
os.environ["OPENAI_API_KEY"] = userdata.get('GROQ_Researcher_API')
os.environ["OPENAI_API_BASE"] = "https://api.groq.com/openai/v1"  # Groq's API endpoint
os.environ["OPENAI_MODEL_NAME"] = "qwen/qwen3-32b"
```

```python
researcher = Agent(
    role='Senior Research Analyst',
    goal='Conduct thorough research on given topics',
    backstory="""You are an experienced research analyst with a keen eye for detail.
    You excel at gathering information from various sources and synthesizing it into
    comprehensive reports.""",
    verbose=True,
    allow_delegation=False,
)

# Create a writer agent
writer = Agent(
    role='Content Writer',
    goal='Create engaging and informative content based on research',
    backstory="""You are a skilled content writer who specializes in transforming
    research findings into clear, engaging, and well-structured content that
    resonates with the target audience.""",
    verbose=True,
    allow_delegation=False,
)
```

```python
research_task = Task(
    description="""Research the latest trends in artificial intelligence and machine learning.
    Focus on recent breakthroughs, industry applications, and future implications.
    Provide a comprehensive overview with key findings.""",
    agent=researcher,
    expected_output="A detailed research report with key AI/ML trends, breakthroughs, and implications"
)

writing_task = Task(
    description="""Based on the research findings, create an engaging blog post about
    AI/ML trends. Make it accessible to a general audience while maintaining technical accuracy.
    Include an introduction, main points, and conclusion.""",
    agent=writer,
    expected_output="A well-structured blog post about AI/ML trends for general audience",
    context=[research_task]  # This task depends on the research task
)
```

```python
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, writing_task],
    process=Process.sequential,
    verbose=True
)
```

```python
result = crew.kickoff()
print("\n" + "="*50)
print("FINAL RESULT:")
print("="*50)
print(result)
```
